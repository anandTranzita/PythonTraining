# Fill in Missing Values with the fillna Method
    .The fillna method replaces missing NaN values with its argument
    .The fillna method is available on both DataFrames and Series
    .An extracted Series is a view on the orginal DataFrame, but the fillna method returns a copy 

```python
import pandas as pd
```

```python
nba = pd.read_csv("C:/Users/sharma/Desktop/pandas/pandas/Incomplete/nba.csv")
nba
```

```python
nba.dropna(how = "all")
```

```python
nba.fillna(0)
nba['Salary'] = nba["Salary"].fillna(0)
```

```python
nba
```

```python
nba['College'] = nba["College"].fillna(value = "Unknown")
```
